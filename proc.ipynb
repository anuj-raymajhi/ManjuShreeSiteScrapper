```python
import pandas as pd
from langchain_groq import ChatGroq
from langchain.prompts import PromptTemplate
```

```python
from dotenv import find_dotenv, load_dotenv
load_dotenv(find_dotenv())
```

```python
llm = ChatGroq(
            model="llama-3.1-70b-versatile",
            temperature=0,
            max_tokens=None,
            timeout=None,
            max_retries=2,
        )
```

```python
prompt = PromptTemplate(
            input_variables=["content"],
            template="""
You are an intelligent data processing bot. Your task is to analyze the provided data, extract key concepts, terms, and entities, and generate highly relevant search tags and metadata that can optimize vector-based search performance. Consider the following when creating the tags:

* Identify the most important and unique terms or phrases in the data.
* Group related keywords and synonyms for comprehensive tagging.
* Consider user intent and contextual relevance.
* Aim to improve search result accuracy, recall, and precision.

Once you've created the tags and metadata, make sure they are well-organized and appropriately weighted to enhance vector search efficiency.

Content to analyze:
{content}
""")
```

```python
chain = prompt | llm
```

```python
data = pd.read_csv('./ManjuShreeSiteData.csv')
```

```python
def proc(val:str)->str:
    response = chain.invoke(input={'content':val})
    # print(val)
    content = f'{val}\n\nTags and Metadata:\n{response.content}'
    return content
```

```python
data['Content'] = data['Content'].apply(proc)
```

```python
for val in data['Content']:
    print(f'\n\n{val}\n\n')
```

```python
data.to_csv('./SiteDataTAG.csv', index=False)
```
